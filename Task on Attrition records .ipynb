```python
import pandas as pd
import sqlite3 as sql
```

```python
database ='''attr_data.db'''
connection = sql.connect(database)
```

```python
query = '''SELECT * FROM attrition_records'''
```

```python
df = pd.read_sql_query(query, connection)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
corr = df.corr() 
```

```python
corr
```

```python
df.describe()
```

```python
#What do you think are the 3 factors behind employee attrition?
```

```python
fig,ax = plt.subplots(1,3, figsize=(20,20))               
plt.suptitle("Distribution of various factors", fontsize=20)
sns.barplot(df['Gender'],df['DistanceFromHome'],hue = df['Attrition'], ax = ax[0]); 
sns.barplot(df['Gender'],df['WorkLifeBalance'],hue = df['Attrition'], ax = ax[1]); 
sns.barplot(df['Gender'],df['NumCompaniesWorked'],hue = df['Attrition'], ax = ax[2]);
```

```python
sns.countplot(x='Department',data=df,hue='Attrition')
```

```python
#It can be observed that the further away an employee is from from the more likey they are to leace the company. Work life balance and number of compnies worked also has a negative impact on attrition can also be seen as main reasons for attrition.
```

```python
#The Attrition Rate is more in the Research and Development Department. whereas the Attrition Rate is very Less in the HR department
```

```python
#What is the relationship between Education and Monthly Income?
```

```python
sns.jointplot(x='Education',y='MonthlyIncome',data=df)
```

```python
#From the plot we can see that the more educated an employee is the highter their income. 
```

```python
#What is the effect of age on attrition?
```

```python
fig, ax = plt.subplots(figsize=(15,20))
sns.countplot(x='Age',data=df,hue='Attrition')
```

```python
#Here we can see that the Attrition rate is more for people between the age of 29 to 31.
```

```python
#Is Income the main factor towards employee attrition?
```

```python
sns.set(style = 'whitegrid')    
       
sns.stripplot(x="Attrition", y="MonthlyIncome", data=df)
  
plt.show()
```

```python
#In this plot we can see that Monthly Income is not a mian factor of attrition as the rate of attrition comparatively low. 
```

```python
#How does work-life balance impact the overall attrition rate?
```

```python
fig, ax = plt.subplots(figsize=(15,20))
sns.countplot(x='WorkLifeBalance',data=df,hue='Attrition')
```

```python
fig, ax = plt.subplots(figsize=(15,20))
sns.barplot(df['Gender'],df['WorkLifeBalance'],hue = df['Attrition'])
```

```python
#As can be seen in the plot, there is a positive relationship between work life balance and attrition rate. This can be seen especially in the females.
```
